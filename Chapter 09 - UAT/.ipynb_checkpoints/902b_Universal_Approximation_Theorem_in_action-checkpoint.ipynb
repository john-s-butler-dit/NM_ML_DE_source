```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import time
```

```python
time1 = time.monotonic_ns()
class Net2(nn.Module):

    def __init__(self, n_input_features):
        super(Net2, self).__init__()
        self.fc1 = nn.Linear(n_input_features, 30)
        self.fc2 = nn.Linear(3, 1)

    def forward(self, x):
        x=            nn.ReLU(),
self.fc1(x)
        x=self.fc2(x)

        return x
```

```python
import matplotlib.pyplot as plt
import numpy as np
torch.manual_seed(5)

x = np.arange(0,7+0.001,1/1000)
y = np.cos(x)
x=x[2:-6]
Y1=y[0:-8]
Y2=y[1:-7]
Y3=y[2:-6]
Y4=y[3:-5]
Y5=y[4:-4]
Y6=y[5:-3]
Y7=y[6:-2]
Y8=y[7:-1]

O=Y4
#O=x*x#np.sin(x)
#I=np.array([x,2*x,2*x,2*x])
I=np.array([x,Y1,Y2,Y3])

I=I.T
I.shape[1]
```

```python
net2=Net2(I.shape[1])
print(net2)
```

```python
x_tor=torch.from_numpy(I.astype(np.float32))
Y_tor=torch.from_numpy(O.reshape(-1,1)).float()
```

```python
num_epochs = 5001
n_plot = np.round(num_epochs/5-1)
learning_rate= 0.01
optimiser= torch.optim.SGD(net2.parameters(),lr=learning_rate)
loss_func=torch.nn.MSELoss()
```

```python
inputs=Variable(x_tor)
outputs=Variable(Y_tor)
Learning=[]
check = 0
i = 0
for i in range(num_epochs):
    prediction = net2(inputs)
    loss = loss_func(prediction, outputs)
    optimiser.zero_grad()
    loss.backward()
    optimiser.step()
    Learning.append(loss.data.numpy())

    if i > 1 and np.absolute(Learning[i] - Learning[i-1]) < 0.00001 and check == 0:
        check = 1
        time2 = time.monotonic_ns()
        print("Critical at Learning Cycle: " + str(int(i)) + ", Learning: " + str(float(Learning[i])))
        print("Time (nanoseconds): " + str(time2-time1))

time3 = time.monotonic_ns()
if check == 0:
    print("Critical Learning not reached")
print("Learning after " + str(i) + " Learning cycles: " + str(float(Learning[i])))
print("Time (nanoseconds): " + str(time3-time1))
```
